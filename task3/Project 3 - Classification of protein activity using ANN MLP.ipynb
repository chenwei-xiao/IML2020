```python
## IPython Commands
%matplotlib inline
# IPython modules
from IPython.display import display

# General math and plotting modules.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# General Setting for pd
pd.set_option('display.max_columns', 100) # Setting pandas to display a N number of columns
pd.set_option('display.max_rows', 10) # Setting pandas to display a N number rows
pd.set_option('display.width', 1000) # Setting pandas dataframe display width to N

# Machine Learning library.
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.DataFrame(pd.read_csv('train.csv'))
test = pd.DataFrame(pd.read_csv('test.csv'))
display(train.shape)
display(train.head())
```

```python
X_features = train.Sequence.apply(lambda x: pd.Series(list(x)),1)
display(X_features.head())
```

```python
# One hot encoding with sklearn: much faster
codes = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
         'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
enc = OneHotEncoder(categories=[codes,codes,codes,codes])
X_features_coded = enc.fit_transform(X_features).toarray()
```

```python
display(pd.DataFrame(X_features_coded).head()) # 80 features
print(X_features_coded.shape)
```

```python
# Our own one hot encoding function
codes = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
         'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']

def one_hot_encode(seq):
    o = list(set(codes) - set(seq))
    s = pd.DataFrame(list(seq))    
    x = pd.DataFrame(np.zeros((len(seq),len(o)),dtype=int),columns=o)    
    a = s[0].str.get_dummies(sep=',')
    a = a.join(x)
    a = a.sort_index(axis=1)
    e = a.values.flatten()
    return e
```

```python
X = train.Sequence.apply(lambda x: pd.Series(one_hot_encode(x)),1)
y = train.Active
display(X.head())
```

```python
# Split dataset into training and validation set
# X is same as X_features_coded
y = np.array(train.Active)
train_split = 0.9
X_train, X_valid, y_train, y_valid = train_test_split(X_features_coded, y, train_size=train_split)
print(F'train/valid size: {len(X_train)}/{len(X_valid)}')
```

```python
# Start to train our Multi-layer Perceptron classifier
# epochs = max_iter = 100
# ReLU activations
# Cross-entropy loss, or log loss default for MLPClassifier

clf = MLPClassifier(max_iter = 100,
                    activation='relu',
                    alpha=0.0001,
                    early_stopping=True,
                    hidden_layer_sizes=(100,),
                    random_state=0, solver='adam')
```

```python
# Another possible classifier form
# clf = MLPClassifier(activation='logistic', #alpha=0.0001,
#                     early_stopping=True)
```

```python
clf.fit(X_train, y_train)
```

```python
print('clf accuracy score for validation set is {}'.format(clf.best_validation_score_)) # very high mean accuracy on the given test data and labels
```

```python
print('f1 score, train:', f1_score(y_train,clf.predict(X_train)))
print('f1 score, valid:', f1_score(y_valid,clf.predict(X_valid)))
```

```python
# Output prediction results
X_test = test.Sequence.apply(lambda x: pd.Series(list(x)),1)
X_test_coded = enc.fit_transform(X_test).toarray()
y_test = clf.predict(X_test)
df = pd.Series(y_test)

df.to_csv('output_n.csv', index=False, header=False)
```
